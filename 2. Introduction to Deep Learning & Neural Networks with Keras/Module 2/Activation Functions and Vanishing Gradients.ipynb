```python
from cProfile import label

import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def sigmoid_derivative(z):
    return sigmoid(z) * (1 - sigmoid(z))
```

```python
def relu(z):
    return np.maximum(0, z)
```

```python
def relu_derivative(z):
    return np.where(z > 0, 1, 0)
```

```python
z = np.linspace(-10, 10, 400)
z
```

```python
sigmoid_grad = sigmoid_derivative(z)
relu_grad = relu_derivative(z)
print(f"Sigmoid grad: {sigmoid_grad}")
print(f"Relu grad: {relu_grad}")
```

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(z, sigmoid(z), label='sigmoid activation', color='blue')
plt.plot(z, sigmoid_grad, label='sigmoid derivative', color='red', linestyle='--')
plt.title('Sigmoid activation and derivative')
plt.xlabel('Input value (z)')
plt.ylabel("Activation/Gradient")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(z, relu(z), label='Relu activation', color='g')
plt.plot(z, relu_grad, label='Relu derivative', color='red', linestyle='--')
plt.title('Relu activation and gradient')
plt.xlabel('Input value (z)')
plt.ylabel('Activation/Gradient')
plt.legend()
plt.show()
```

```python
def tanh(z):
    return np.tanh(z)
def tanh_derivative(z):
    return 1 - np.tanh(z) ** 2
```

```python
z = np.linspace(-5, 5, 100)

tanh_grad = tanh_derivative(z)
relu_grad = relu_derivative(z)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(z, relu(z), label='ReLU Activation', color='g')
plt.plot(z, relu_grad, label="ReLU Derivative", color='r', linestyle='--')
plt.title('ReLU Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(z, tanh(z), label='tanH Activation', color='g')
plt.plot(z, tanh_grad, label="tanH Derivative", color='r', linestyle='--')
plt.title('tanH Activation & Gradient')
plt.xlabel('Input Value (z)')
plt.ylabel('Activation / Gradient')
plt.legend()

plt.tight_layout()
plt.show()
```
