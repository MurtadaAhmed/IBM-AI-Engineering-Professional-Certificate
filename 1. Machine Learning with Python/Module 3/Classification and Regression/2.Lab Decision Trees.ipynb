```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn import metrics
import warnings
warnings.filterwarnings('ignore')
```

```python
file = "drug200.csv"
my_data = pd.read_csv(file)
my_data
```

```python
my_data.info()
```

```python
label_encoder = LabelEncoder()
my_data['Sex'] = label_encoder.fit_transform(my_data['Sex'])
my_data['BP'] = label_encoder.fit_transform(my_data['BP'])
my_data['Cholesterol'] = label_encoder.fit_transform(my_data['Cholesterol'])
my_data
```

```python
my_data.isnull().sum()
```

```python
custom_map = {"drugA": 0, "drugB": 1, "drugC": 2, "drugX": 3, "drugY": 4}
my_data['Drug_num'] = my_data['Drug'].map(custom_map)
my_data
```

```python
my_data.drop("Drug", axis=1).corr()['Drug_num']
```

```python
category_count = my_data['Drug'].value_counts()
category_count
```

```python
plt.bar(category_count.index, category_count.values, color='blue')
plt.xlabel('Drug')
plt.ylabel('Count')
plt.title("Category Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
X = my_data.drop(["Drug", "Drug_num"], axis=1)
y = my_data["Drug"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=32)
```

```python
drugTree = DecisionTreeClassifier(criterion='entropy', max_depth=4)
drugTree.fit(X_train, y_train)
```

```python
tree_prediction = drugTree.predict(X_test)
print("Decision Tree's Accuracy", metrics.accuracy_score(y_test, tree_prediction))
```

```python
plot_tree(drugTree)
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plot_tree(drugTree, feature_names=["Age", "Sex", "BP", "Cholesterol", "Na_to_K"], class_names=drugTree.classes_, filled=True)
plt.savefig("drug_tree.png", dpi=300)
plt.show()
```

```python
drugTree = DecisionTreeClassifier(criterion="entropy", max_depth=3)
drugTree.fit(X_train, y_train)
tree_prediction = drugTree.predict(X_test)
print(metrics.accuracy_score(y_test, tree_prediction))
```

```python
plot_tree(drugTree)
plt.show()
```
